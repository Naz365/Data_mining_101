```python
import pandas as pd
```

```python
with open ('car.data') as carFile:
  data=carFile.read()
  data=data.split('\n')
  
```

```python
print(data)
```

```python
newCardata=[]
for i in data:
  newCardata.append(i.split(','))
```

```python
newCardata
```

```python
newCardata[0][1]
```

```python
df=pd.DataFrame(newCardata,columns=["buying","maint","doors","persons","lug_boot","safety","class"])
```

```python
df.to_csv("carData",index=False)
```
