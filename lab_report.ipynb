```python
import pandas as pd
import numpy as np
data = pd.read_csv("train.csv")
print(data.head())
```

```python
# Step 2: Explore the data
print("Data Overview:")
print(data.head())
print("\nData Types:")
print(data.dtypes)
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Step 3: Data Cleaning
data.info()

# Drop unnecessary columns or columns with high missing values
data_cleaned = data.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1)

# Fill missing values for 'Age' with mean value
data_cleaned['Age'].fillna(data_cleaned['Age'].mean(), inplace=True)

# Fill missing values for 'Embarked' with the most common value
data_cleaned['Embarked'].fillna(data_cleaned['Embarked'].mode()[0], inplace=True)
```

```python
# Impute missing values in 'Age' with median value
data['Age'].fillna(data['Age'].median(), inplace=True)

# Impute missing values in 'Embarked' with the most common value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
# Data transformation

# 1. Convert 'Sex' column to numeric (0 for female, 1 for male)
data['Sex'] = data['Sex'].map({'female': 0, 'male': 1})

# 2. Encode 'Embarked' column using one-hot encoding
data_encoded = pd.get_dummies(data, columns=['Embarked'])

# 3. Create a new feature 'FamilySize' by combining 'SibSp' and 'Parch'
data_encoded['FamilySize'] = data_encoded['SibSp'] + data_encoded['Parch'] + 1

# 4. Binning 'Age' column into age groups
age_bins = [0, 12, 18, 30, 50, 100]
age_labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
data_encoded['AgeGroup'] = pd.cut(data_encoded['Age'], bins=age_bins, labels=age_labels)
print(data_encoded)
```
