```python

import pandas as pd

pima = pd.read_excel('loan.xlsx')

pima.head()
```

```python
feature_cols = ['age', 'annual_income','credit_score','loan_amount', 'previous_payments', 'default']
x = pima[feature_cols]
y = pima.default
```

```python
pima.fillna(pima.mean(), inplace=True)
pima.drop_duplicates(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state
=0)
```

```python
from sklearn.linear_model import LogisticRegression


logisticRegr = LogisticRegression()

```

```python
logisticRegr.fit(x_train,y_train)
y_pred = logisticRegr.predict(x_test).astype(int)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
class_names=[0,1] # name of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu",fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
